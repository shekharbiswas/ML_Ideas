```python
# Simplified Df to run last analysis on customer behaviour and use of bikes in the different seasons

y2019.drop(columns=["Start station number","Start station","End station number","End station","End date", "Bike number", "Duration"], inplace=True)
y2019.rename(columns={"Start date" : "Start_date", "Member type": "Member_type"},inplace=True)
y2019.head()
```

```python
# extract hour, date and month from 2019 data and created new columns in df to be used for visualization 

y2019["Start_date"] = pd.to_datetime(y2019["Start_date"])

y2019['date'] = y2019['Start_date'].dt.date
y2019['time'] = y2019['Start_date'].dt.hour
y2019["month"]= y2019['Start_date'].dt.month
```

```python
# Creation of Feature seasons using lambda function

y2019['season'] = y2019['month'].apply(lambda value: 'Winter'
if value <= 3 else 'Spring'
if value <= 6 else 'Summer'
if value <= 9 else "Fall")
y2019['season'] = pd.Categorical(y2019['season'],categories=["Winter", "Spring", "Summer", "Fall"])
y2019.head()
```

```python
# Plot with season trend - count of rides on y axis, seasons on x axis

y2019["season"].value_counts().plot(kind="line").yaxis.get_major_formatter().set_scientific(False)
```

```python
time_seasons = y2019.groupby(["time","season"]).size()
time_seasons = time_seasons.reset_index()
time_seasons.columns=["time", "season", "count"]
time_seasons.head()
```

```python
# Distribution of the rides over the different time of the day and the different seasons

sns.set(rc={'figure.figsize':(11,8)})
sns.set_context("paper", font_scale=1, rc={"lines.linewidth": 2.5})
sns.set_style("dark")

ax = sns.pointplot(data = time_seasons[['time','count','season']], x = 'time', y = 'count',
              scale = 0.5, hue = 'season');

plt.setp(ax.get_legend().get_texts(), fontsize='15')
plt.setp(ax.get_legend().get_title(), fontsize='16')
plt.title('Season wise hourly distribution of bike rentals(2018-19)',fontsize="xx-large")
plt.ylabel('Count',fontsize="large")
plt.xlabel("Time", fontsize="large")
plt.tight_layout()
plt.savefig('Season wise hourly distribution of bike rentals(2019).png', dpi = 80)
```
